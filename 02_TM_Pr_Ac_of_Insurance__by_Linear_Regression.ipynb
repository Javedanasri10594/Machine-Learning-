```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data['sex']=data['sex'].apply({'male':0,'female':1}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southeast':1,'southwest':2,'northeast':3,'northwest':4}.get)
```

```python
data.head()
```

```python
x=data[["age","sex","bmi","children","smoker","region"]]
y=data[["expenses"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
pridication=regression.predict(x_test)
```

```python
regression.score(x,y)
```
