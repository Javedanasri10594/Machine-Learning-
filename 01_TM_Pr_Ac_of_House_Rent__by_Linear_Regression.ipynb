#(1)Gathering the Data

```python
#Impoarting the Necessary Packages
import numpy as np
import pandas as pd
```

#(2) Data Preprocessing

```python
# convert the categorical variable into numerical formet - using the Hot  Encoding Technique
data['sex']=data['sex'].apply({'male':0,'female':1}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southeast':1,'southwest':2,'northeast':3,'northwest':4}.get)
```

```python
data.head()
```

#(3)Dividing the Clean data into
#(X)-Dependent
#(y)-Independent Veariable

```python
x=data[['age','sex','bmi','children','smoker','region']] #Independent Parameter
y=data['expenses'] #Dependent Parameter
```

#(4)Spliting the data into training and testing set


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

Creating the Machine Learning model(MLM) using LinerRegression Algorithm

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()  # Activeatring the Linear Regressiobn Algorithm
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train) #.fit() used for the training the MLM
```

```python
pridication=regression.predict(x_test) # Pridicat() used to predict the feture Value
```

```python
# Accurecy for MLM
regression.score(x,y)
```
